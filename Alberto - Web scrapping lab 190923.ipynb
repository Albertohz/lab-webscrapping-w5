````python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    # find all img tags
    img_tags = soup.find_all('img')

    # getting the links
    image_links = [img.get('src') for img in img_tags]

    # Filterig=ng out None values
    image_links = list(filter(None, image_links))

    for link in image_links:
        print(link)
else:
    print('Failed to retrieve the page.')
````

````python
soup = BeautifulSoup(response.content)
soup
````

````python
# as before with developpers

titles = []
for component in soup.find_all('h3', class_='ipc-title__text'):
    title = component.text
    titles.append(title)

titles = titles[1:251]
titles

initial_release = []
for component in soup.find_all('span', class_='sc-4dcdad14-8 cvucyi cli-title-metadata-item'):
    date = component.text
    initial_release.append(date)

initial_release = initial_release[::3]
initial_release

````

````python
soup.find('span', class_='ipc-rating-star').text.split('\xa0')[0]
ratings = []
for component in soup.find_all('span', class_='ipc-rating-star'):
    rating = component.text.split('\xa0')[0]
    ratings.append(rating)

ratings = ratings[::2]
ratings
````

````python
titles_sr= pd.Series(titles)
initial_release_sr = pd.Series(initial_release)
ratings_sr= pd.Series(ratings)

best_movies =pd.concat([titles_sr, initial_release_sr, ratings_sr], axis=1)
````

````python
best_movies.rename(columns={0:'Title', 1:'Date', 2:'Rating'}, inplace=True)
best_movies
````
